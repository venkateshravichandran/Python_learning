# Pandas Moduel

Pandas Dataframe can be created in two ways: Using Array and Dictionary

```python
import pandas as pd
import numpy as np
```

## 1. Creating Dataframe using Array

```python
# 2D array created
two_D_array = np.array([[1,4],[2,5],[3,6]])

# Creating Dataframe from Array
# Column name and row name can change by proving the column and index label
myDataframe_on_array =pd.DataFrame(two_D_array,index=['row1','row2','row3'],columns=['col1','col2'])

myDataframe_on_array
```

## 2. Creating Dataframe using Dictionary

```python
# dictionay created
myDict = {'col1':[1,2,3],'col2':[4,5,6]}

# Creating Dataframe using dictionary

# Here the column label was created. the pandas methods fetched from dictionry key as column label
myDataframe_on_dict = pd.DataFrame(myDict,index=['row1','row2','row3'])

myDataframe_on_dict
```

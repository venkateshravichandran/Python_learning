## 1. Line Plot
When you call plot() method directly on a DataFrame, by default, the method will plot each column of the DataFrame as a separate series. Each column becomes a line plot on the same set of axes. This behavior is useful when you have multiple series of data in different columns that you want to visualize together.
This method contains many prameters as follow:
- kind = 'line' # graph type (bar,pie,box)
- figsize=(12,3) # tuples(X-axis size, y-axis size)
- xlabel='Date'
- ylabel='NDVI'
- title='Line Chart'
- ylim=(50,180) # tuples(y-axis lower limit, y-axis upper limit)

```python
new_DataFrame = plot_gdf.iloc[0:10,0:5]
new_DataFrame.plot(kind = 'line',figsize=(12,4), xlabel='Date',ylabel='NDVI', title='Line Chart', ylim=(0,180))
```

```python
# Removing values below 99 from each row of columns and ploting the graphs.
plt.figure(figsize=(20, 6))
for i,row in plot_gdf[0:10].iterrows():
    line_row = []
    for val in row:
        if val >99:
            line_row.append(val)
        plt.plot(line_row,marker='o')
plt.ylim(100,180)
plt.show()
```

```python
# Same operation using List comprehension
plt.figure(figsize=(20, 6))
for i,row in plot_gdf[0:10].iterrows():
    line_row = [val for val in row if val>99]
    plt.plot(line_row,marker='o')
plt.ylim(100,180)
plt.show()
```

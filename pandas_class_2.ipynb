### 1.2 Selecting one column using '.' notation (syntax-2)

## 2 Adding new column in DataFrame

### 2.1 Adding new column with the scalar value
- This might be the most commonly used method for creating a new column.

```python
# adding new column to DataFrame using [] bracket similar to accessing the column by assinging name of the columns.
# assigning a scalar value (default) for all the rows

df['per_area'] = 0
df
```

### 2.2 Adding new column with insert () method
- Pandas also allows for adding new columns at a specific index. The insert function can be used to customize the location of the new column.

```python
df.insert(len(df.columns),'temp_col',2)
df
```

### 2.3 Adding new column with loc method
- The loc method allows you to select rows and columns using their labels. It’s also possible to create a new column with this method.

```python
df.loc[:,'temp_col5'] = 0
df
```

### 2.4 Adding new column with assigning a function 
- The Dataframe.assign() method assigns new columns to a DataFrame, returning a new object (a copy) with the new columns added to the original ones.


```python
# If the values are callable, they are computed on the DataFrame and assigned to the new columns.
# If the values are not callable, (e.g. a Series, scalar, or array), they are simply assigned as value.
# Here the function name will be assigned to column label. "x" represents the DataFrame
new_df = df.assign(new_col = lambda x: df['polygon_area'])
new_df = df.assign(new_col = lambda x: df['polygon_area']/10)
new_df
```

## 3. Operation in DataFrame

```python
# select the column and caculate total sum
print(df['polygon_area'].sum())
print(df['polygon_area'].count())
print(df['polygon_area'].mean())
print(df['polygon_area'].std())
print(df['polygon_area'].max())
print(df['polygon_area'].min())
```

## 5. Sort a DataFrame with the sort_value() method

```python
# sorting the single column by sort_value method. By default the values are sorted in asceding order. We can change by assigning False to ascending
df.sort_values('polygon_area')
```

```python
df.sort_values('polygon_area',ascending=False) # value are sorted decending order
```

```python
# sorting multiple columns by sort_values methods
df.sort_values(['polygon_area','validation']) # both columns are sorted ascending order
```

```python
#Still now the sort_values method creates an copy of DataFrame and modify accordingly. By changing the inplace values into True, so that modified
# values will be updated into DataFrame
df.sort_values(['polygon_area'],ascending=False,inplace=True)
df # now the new modified values updated into DataFrame with decending order
```

### (%) map() method

```python
#Dataframe.map() method applies a function that accepts and returns a scalar to every element of a DataFrame.
# This method walk through all individual elements of the DataFrame and apply the function.This method will return an modified DataFrame
# This method used to apply some changes for all the elements of the DatFrame.

df.map(lambda x:len(str(x)))
```

### (%) apply() method

```python
#DataFrame.apply() method apply a function along an axis of the DataFrame. Either row-wise or column-wise.
# This method walk through all individual elements of the DataFrame and apply the function.This method will return an modified DataFrame
# This method used to apply some changes for all the elements of the DatFrame.
df.apply(lambda x:x.sum(),axis=0)[0]
```

## 6. DataFrame Tranformation

### 6.1 Pivot () method

```python
# Return reshaped DataFrame organized by given index / column values. This is used only for data tranformation. If the duplicate values presents, it 
# throws an error. Insuch case, we can switch to pivot_table() method to aggregate the duplicate values

df.pivot(index='polygon_id',columns='crop_name')
```

### 6.2 pivot_table() method

```python
# Return reshaped DataFrame organized by index, column/values with aggregate method. 
df.pivot_table(index='user_name',columns='crop_name',values="polygon_id",aggfunc='count')
```

```python
dff = pd.pivot_table(df,index='user_name',columns='crop_name',values="polygon_id",aggfunc='count')
dff
```

## 7. Slicing methods
- `iloc()` and `loc()` method

```python
# These are used in slicing data from the Pandas DataFrame. They help in the convenient selection of data from the DataFrame in Python. 
# They are used in filtering the data according to some conditions.
```

```python
# iloc() methods gets two agruments. row slicing and column slicing separated by commas ",".
# The following slicing shows as 0 to 5 index of rows and 0 to 2 index of columns 
df.iloc[0:5,0:2]
```

```python
# this will give row of index 5 to 10 and all rows, bcz we did'nt speficied any column slicing
df.iloc[5:8]
```

```python
# Access a group of rows and columns by label(s) or a boolean array.
#.loc[] is primarily label based, but may also be used with a boolean array
# We can slice row and columns based on index or boolean values. similar to iloc() method, Access a group of rows and columns by label(s) or a boolean array.
# Below code, only row filter according to the condition. If the conditiono satisfy it returns the respective rows.

df.loc[df['polygon_area']>2500] # Conditional that returns a boolean Series
```

```python
# Addition to previous code, filter the required columns 
df.loc[df['polygon_area']>2500, ['collection_date','state','district','village','crop_name']] 
#Conditional that returns a boolean Series with column labels specified
```

```python
# Addition to previous code, for row filtering, boolean 
df.loc[(df['polygon_area']>2500) & (df['district']=='Ahmednagar'), ['collection_date','state','district','village','crop_name']]
# Multiple conditional using & that returns a boolean Series
```
